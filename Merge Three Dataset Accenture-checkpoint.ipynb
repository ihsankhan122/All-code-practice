Overall Goal:
To merge three different datasets into one comprehensive dataset and then determine the top 5 performing categories based on the total scores of reactions.

```python

import pandas as pd

# Load data into pandas DataFrames
reaction_df = pd.read_csv('Reactions.csv')
content_df = pd.read_csv('Content.csv')
reaction_types_df = pd.read_csv('ReactionTypes.csv')
```

```python
# Display column names and first few rows of each DataFrame
print(reaction_df.columns)
print(reaction_df.head())

print(content_df.columns)
print(content_df.head())

print(reaction_types_df.columns)
print(reaction_types_df.head())
```

```python
# Ensure that 'Type' column exists in both DataFrames
print("Columns in merged_df after first merge:")
merged_df_temp = pd.merge(reaction_df, content_df, on='Content ID', how='left')
print(merged_df_temp.columns)
```

```python
print("Columns in reaction_types_df:")
print(reaction_types_df.columns)
```

```python
# Ensure 'Type' column exists in merged_df
if 'Type_x' in merged_df.columns:
    merged_df.rename(columns={'Type_x': 'Type'}, inplace=True)
elif 'Type' not in merged_df.columns:
    raise KeyError("'Type' column is missing after the first merge.")
```

```python
print("Missing values in 'Type' column of merged_df:", merged_df['Type'].isnull().sum())
```

```python

# Drop rows with missing 'Type' values in merged_df
merged_df.dropna(subset=['Type'], inplace=True)
```

```python
# Merge with reaction_types_df on 'Type'
merged_df = pd.merge(merged_df, reaction_types_df, on='Type', how='left')
print("Columns in merged_df after second merge:", merged_df.columns)
```

```python
# Select only the columns you need in the final merged DataFrame
final_df = merged_df[['Content ID', 'User ID_x', 'Type', 'Datetime', 'Score', 'Category', 'Sentiment']]
```

```python

# Optionally, clean and process data further as needed
# Example: Handle missing values, data type conversions, etc.
final_df.dropna(inplace=True)  # Drop rows with any NaN values
final_df.reset_index(drop=True, inplace=True)  # Reset index if needed
```

```python
# Display the final merged DataFrame
print(final_df.head())
```

```python
# Calculate total scores for each category
category_scores = final_df.groupby('Category')['Score'].sum().reset_index()
```

```python
category_scores
```

```python
# Sort categories by total score in descending order
category_scores = category_scores.sort_values(by='Score', ascending=False)
```

```python
category_scores
```

```python
# Get top 5 performing categories
top_5_categories = category_scores.head(5)
```
