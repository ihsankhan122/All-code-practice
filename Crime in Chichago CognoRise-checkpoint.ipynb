```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("Chicago_Crimes_2001_to_2004.csv")
df
```

```python
df.head()
```

```python
# Check the structure and contents of the dataset
print("Dataset Info:")
df.info()
```

```python
# Explore summary statistics of numerical variables
print("\nSummary Statistics of Numerical Variables:")
df.describe()
```

```python
 #Fill missing values with 0
df.fillna(0, inplace=True)

```

```python
# Check if missing values are replaced with 0
print("\nMissing Values after replacing with 0:")
print(df.isnull().sum())
```

```python
df.dtypes
```

```python
# Deal with duplicates
df.drop_duplicates(inplace=True)
```

```python
# Convert data types if necessary
# Convert date columns to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Explore the distribution of numerical variables
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Beat', y='Year', bins=20, kde=True)  # Example with 'Beat' on x-axis and 'Year' on y-axis
plt.title('Distribution of Crime Incidents Across Beats')
plt.xlabel('Beat')
plt.ylabel('Year')
plt.show()

```

```python
loc_to_change  = list(df['Location Description'].value_counts()[20:].index)
desc_to_change = list(df['Description'].value_counts()[20:].index)
```

```python
df.loc[df['Location Description'].isin(loc_to_change) , df.columns=='Location Description'] = 'OTHER'
df.loc[df['Description'].isin(desc_to_change) , df.columns=='Description'] = 'OTHER'
#crimes.loc[crimes['Primary Type'].isin(type_to_change) , crimes.columns=='Primary Type'] = 'OTHER'
```

```python
# we convert those 3 columns into 'Categorical' types -- works like 'factor' in R
df['Primary Type']         = pd.Categorical(df['Primary Type'])
df['Location Description'] = pd.Categorical(df['Location Description'])
df['Description']          = pd.Categorical(df['Description'])
```

```python


# Set 'Date' column as the index
df.set_index('Date', inplace=True)

# Now you can use the resample method
plt.figure(figsize=(11,5))
df.resample('M').size().plot(legend=False)
plt.title('Number of crimes per month (2001 - 2004)')
plt.xlabel('Months')
plt.ylabel('Number of crimes')
plt.show()
```

```python
plt.figure(figsize=(11,4))
df.resample('D').size().rolling(365).sum().plot()
plt.title('Rolling sum of all crimes from 2001 - 2004')
plt.ylabel('Number of crimes')
plt.xlabel('Days')
plt.show()
```

```python
crimes_count_date = df.pivot_table('ID', aggfunc=np.size, columns='Primary Type', index=df.index.date, fill_value=0)
crimes_count_date.index = pd.DatetimeIndex(crimes_count_date.index)
plo = crimes_count_date.rolling(365).sum().plot(figsize=(12, 30), subplots=True, layout=(-1, 3), sharex=False, sharey=False)
```

```python
df.groupby([df.index.month]).size().plot(kind='barh')
plt.ylabel('Months of the year')
plt.xlabel('Number of crimes')
plt.title('Number of crimes by month of the year')
plt.show()
```

```python
# Compare crime rates across different days of the week
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x=df.index.dayofweek)
plt.title('Number of Crimes by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crimes')
plt.xticks(ticks=range(7), labels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.show()
```

```python
# Compare crime rates across different times of the day
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x=df.index.hour, bins=24, kde=True)
plt.title('Number of Crimes by Time of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
plt.figure(figsize=(8,10))
df.groupby([df['Primary Type']]).size().sort_values(ascending=True).plot(kind='barh')
plt.title('Number of crimes by type')
plt.ylabel('Crime Type')
plt.xlabel('Number of crimes')
plt.show()
```

```python
plt.figure(figsize=(8,10))
df.groupby([df['Location Description']]).size().sort_values(ascending=True).plot(kind='barh')
plt.title('Number of crimes by Location')
plt.ylabel('Crime Location')
plt.xlabel('Number of crimes')
plt.show()
```
