# LINE CHART

```python
import matplotlib.pyplot as plt

# Sample data
x_values = [1, 2, 3, 4, 5]
y_values = [10, 8, 12, 6, 14]

# Create the line chart
plt.plot(x_values, y_values, marker='o', linestyle='-', color='blue')

# Customize the chart
plt.title('Line Chart Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the chart
plt.show()
```

# Area and Stacked Plots

```python
import matplotlib.pyplot as plt

# Sample data
x_values = [1, 2, 3, 4, 5]
y1_values = [10, 15, 12, 8, 14]
y2_values = [8, 12, 10, 6, 11]

# Create the area plot
plt.stackplot(x_values, y1_values, y2_values, labels=['Y1', 'Y2'])

# Customize the chart
plt.title('Area Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x_values = [1, 2, 3, 4, 5]
y1_values = [10, 15, 12, 8, 14]
y2_values = [8, 12, 10, 6, 11]

# Create the stacked plot
plt.plot(x_values, y1_values, label='Y1')
plt.plot(x_values, y2_values, label='Y2')
plt.fill_between(x_values, y1_values, y2_values, alpha=0.5)

# Customize the chart
plt.title('Stacked Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [30, 20, 25, 15]
explode = (0.1, 0, 0, 0)
colors = ['red', 'green', 'blue', 'yellow']

# Create the pie chart
plt.pie(values, labels=categories, explode=explode, colors=colors, autopct='%1.1f%%', shadow=True)

# Customize the chart
plt.title('Pie Chart Example')

# Display the chart
plt.show()
```

# Polar Charts

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [0.6, 0.8, 0.7, 0.9, 0.5]

# Convert the categories to angles
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

# Convert the values to close the plot
values += values[:1]

# Create the polar chart
plt.polar(angles, values, marker='o')

# Customize the chart
plt.xticks(angles[:-1], categories)
plt.title('Simple Polar Chart Example')

# Display the chart
plt.show()
```

# Histograms

```python
import matplotlib.pyplot as plt

# Sample data
data = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]

# Create the histogram
plt.hist(data, bins=5)

# Customize the chart
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the chart
plt.show()
```

# Lollipop chart

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [10, 20, 15, 25]

# Create the lollipop chart
plt.stem(categories, values, markerfmt='o', linefmt='-', basefmt=' ')

# Customize the chart
plt.title('Lollipop Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')

# Display the chart
plt.show()
```
