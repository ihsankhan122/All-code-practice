```python
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
```

```python
from sklearn.model_selection import train_test_split
X_tran,X_test,y_train,y_test=train_test_split(cancer['data'],cancer['target'])
```

```python
from sklearn.linear_model import Ridge
r=Ridge()

r.fit(X_tran,y_train)
print("traing set",r.score(X_tran,y_train))
r.predict(X_test)
print("test set",r.score(X_test,y_test))
```
