```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("salaries.csv")
df
```

```python
df.head()
```

```python
# Data Cleaning and Preprocessing
# 1. Check for Missing Values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# 2. Handle Missing Values
df.dropna(inplace=True)
```

```python
# 3. Check for Duplicates
duplicates = df.duplicated().sum()
print("Duplicates:", duplicates)
```

```python
4. #Handle Duplicates
df.drop_duplicates(inplace=True)
```

```python
# 5. Ensure Consistency in Categorical Variables
df['experience_level'] = df['experience_level'].str.upper()
df['employment_type'] = df['employment_type'].str.upper()
```

```python

# 6. Convert Data Types if Necessary
df['work_year'] = pd.to_datetime(df['work_year'], format='%Y')
```

```python
# Data Exploration
# Numerical Variables
numerical_vars = ['salary', 'remote_ratio']  # Add other numerical variables if needed
```

```python
df[numerical_vars].describe()
```

```python
# Mapping of abbreviations to full names for experience_level
experience_level_mapping = {
    'EN': 'Entry-Level',
    'MI': 'Mid-Level',
    'SE': 'Senior-Level',
    'EX': 'Executive-Level'
}

# Mapping of abbreviations to full names for employment_type
employment_type_mapping = {
    'PT': 'Part-Time',
    'FT': 'Full-Time',
    'CT': 'Contract',
    'FL': 'Freelance'
}
```

```python
# Replace abbreviations with full names in the dataframe
df['experience_level'] = df['experience_level'].map(experience_level_mapping)
df['employment_type'] = df['employment_type'].map(employment_type_mapping)
```

```python
# Categorical Variables
categorical_vars = ['experience_level', 'employment_type']  # Add other categorical variables if needed
for var in categorical_vars:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=var)
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.show()
```

```python
# Correlation Analysis
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Feature Engineering (if necessary)

# Analysis by Work Year
# Trends in Salary Over Different Years
avg_salary_by_year = df.groupby('work_year')['salary'].mean()
plt.figure(figsize=(10, 6))
avg_salary_by_year.plot(marker='o')
plt.title('Average Salary Over Years')
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.show()
```

```python
# Compare Average Salaries Across Different Categories for Each Year
for var in ['experience_level', 'employment_type', 'company_size']:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df, x='work_year', y='salary', hue=var)
    plt.title(f'Average Salary by {var} Over Years')
    plt.xlabel('Year')
    plt.ylabel('Average Salary')
    plt.legend(title=var)
    plt.show()
```

```python
 #Compare Average Salaries Across Different Categories for Each Year
for var in ['experience_level', 'employment_type', 'company_size']:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df, x='work_year', y='salary', hue=var)
    plt.title(f'Average Salary by {var} Over Years')
    plt.xlabel('Year')
    plt.ylabel('Average Salary')
    plt.legend(title=var, loc='upper left')  # Adjust legend location if needed
    plt.xticks(rotation=45)  # Rotate x-axis labels if necessary for better readability
    plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='salary', y='company_location', hue='salary', size='salary', palette='viridis')
plt.title('Salary Distribution by Company Location (Geographical Analysis)')
plt.xlabel('employee_residence')
plt.ylabel('company_location')
plt.legend(title='Salary')
plt.show()
```
